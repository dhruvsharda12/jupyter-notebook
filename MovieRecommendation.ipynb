```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader)

trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

model = SVD()
model.fit(trainset)

predictions = model.test(testset)
accuracy.rmse(predictions)
```

```python
def recommend_movies(user_id, n=5):
    
    # Movies already watched by user
    watched_movies = ratings[ratings['userId'] == user_id]['movieId'].tolist()
    
    all_movies = movies['movieId'].unique()
    predictions = []

    for movie_id in all_movies:
        if movie_id not in watched_movies:
            pred = model.predict(user_id, movie_id)
            predictions.append((movie_id, pred.est))

    predictions.sort(key=lambda x: x[1], reverse=True)
    top_movies = predictions[:n]

    print(f"\nTop {n} Recommended Movies for User {user_id}:\n")

    for movie_id, rating in top_movies:
        title = movies[movies['movieId']==movie_id]['title'].values[0]
        print(f"{title} â†’ Predicted Rating: {round(rating,2)}")
```

```python
recommend_movies(1)
```
