```python
from sklearn.datasets import load_digits

```

```python
datasets = load_digits()
x = datasets.data
y = datasets.target
```

```python
x.shape
```

```python
y.shape
```

```python
#Scaling on the given datasets
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(x)
```

```python
#apply pca to reduce dimension
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=61)                                          # fro doing directly ,or counting no of features required ->> you can put pca = PCA(0.95)
pca_x_scaled = pca.fit_transform(x_scaler)
```

```python
pca_x_scaled.shape
```

```python
pca.explained_variance_ #eigen values
```

```python
pca.explained_variance_ratio_
```

```python
import numpy as np
arr = np.cumsum(pca.explained_variance_ratio_)  # cumulative frequency
arr
```

```python
np.argmax(arr>0.95)+1  #39th index per 95 hoga to plus 1
```
